本章將介紹 Python 的流程控制，包括決策結構與迴圈結構，並輔以範例程式碼與註解，幫助學生理解邏輯流程。

## 4-1 認識流程控制
Python 的流程控制分為兩大類：
- **決策結構 (if)**：根據條件式為 True 或 False 來決定要執行的程式碼。
- **迴圈結構 (for / while)**：重複執行特定程式碼。

在 Python 中，以下的值會被視為 False，其餘皆為 True：`None`, `False`, `0`, `0.0`, `0j`, 空字串、空串列、空集合、空字典。

## 4-2 if 敘述
### 4-2-1 單向 if
只在條件為 True 時執行內容。

```python
x = 15
y = 10

if x > y:
    z = x - y
    print("x 比 y 大", z)  # 印出差值
```

### 4-2-2 雙向 if...else
當條件不成立時，執行 else 部分。

```python
score = eval(input("請輸入數學分數 (0 ~ 100)："))
if score >= 60:
    print("及格！")
else:
    print("不及格！")
```

### 4-2-3 多向 if...elif...else
可以檢查多個條件。

```python
score = eval(input("請輸入數學分數 (0 ~ 100)："))
if score >= 90:
    print("優等")
elif score >= 80:
    print("甲等")
elif score >= 70:
    print("乙等")
elif score >= 60:
    print("丙等")
else:
    print("不及格")
```

### 4-2-4 巢狀 if
if 中還可以再包含 if，形成巢狀結構。

```python
score = eval(input("請輸入數學分數 (0 ~ 100)："))
if score >= 90:
    print("優等")
else:
    if score >= 80:
        print("甲等")
    else:
        if score >= 70:
            print("乙等")
        else:
            if score >= 60:
                print("丙等")
            else:
                print("不及格")
```

## 4-3 for 迴圈
`for` 迴圈用來針對可迭代物件進行重複運算，例如 range、串列或字串。

### 使用 range 物件

```python
print(list(range(5)))        # 0~4
```

```python
print(list(range(1, 10, 2)))    # 間隔為 2
```

```python
print(list(range(10, -10, -2)))  # 遞減
```

### 使用 range 搭配 for

```python
for i in range(5):
    print(i)  # 依序印出 0~4
```

### 使用串列

```python
list1 = [15, 20, 33, 7, 8]
sum_val = 0
for i in list1:
    sum_val += i
print("總和等於", sum_val)
```

### 使用字串

```python
str1 = "Hello, World!"
for i in str1:
    print(i, '-', end='')
```

### 巢狀 for (九九乘法表)

```python
result = ''
for i in range(1, 10):
    for j in range(1, 10):
        result += f"{i}*{j}={i*j}\t"
    result += '\n'
print(result)
```

## 4-4 while 迴圈
只要條件成立，就會一直執行。

```python
i = 0
while i < 5:
    print(i)
    i += 1  # 遞增避免無限迴圈
```

### while...else 範例

```python
answer = input("請輸入「快樂」的英文：")
while answer.upper() != "HAPPY":
    answer = input("答錯了，請重新輸入「快樂」的英文：")
else:
    print("答對了！")
```

## 4-5 break 與 continue 敘述
- **break**：強制跳出迴圈。
- **continue**：跳過當次迴圈，直接進入下一次檢查。

```python
answer = input("請輸入「快樂」的英文：")
while answer.upper() != "HAPPY":
    if answer.upper() == "QUIT":
        print("我不玩了！")
        break
    answer = input("答錯了，請重新輸入「快樂」的英文：")
else:
    print("答對了！")
```

```python
i = 0
while i < 10:
    i += 1
    if i % 3 != 0:
        continue  # 跳過非 3 的倍數
    print(i)
```
